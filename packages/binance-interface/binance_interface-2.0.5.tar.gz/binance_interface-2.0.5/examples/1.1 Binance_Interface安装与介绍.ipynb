## 1. Binance_Interface安装

pip3 install binance_interface --upgrade

## 2. Binance_Interface介绍

Binance_Interface包含对Rest API接口的完全封装与APP高级封装，并长期更新维护，保证框架的有效性。

**框架结构**

Binance_Interface

- api 官方原始接口封装
    - SPOT 现货
        - SPOTAccount 账户接口
        - SPOTTrade 交易接口
        - SPOTMarket 行情接口
        - SPOTAlgo 策略交易接口
    - margin 杠杆
        - MarginAccountTrade 杠杆账户和交易接口
        - MarginBlvt 杠杆代币接口
    - um U本位合约
        - UMAccountTrade 账户和交易接口
        - UMMarket 行情接口
        - UMPortfolio 经典统一账户
    - cm 币本位合约
        - CMAccountTrade 账户和交易接口
        - CMMarket 行情接口
        - CMPortfolio 经典统一账户接口
    - eo 欧式期权
        - EOAccountTrade 账户和交易接口
        - EOMarket 行情接口
        - EOMarketMaker 做市商接口
    - pm 统一账户
        - PMAccount 账户信息
        - PMMarket 行情接口
        - PMGetOrder 查询订单
        - PMSetOrder 下单接口
        - PMCancelOrder 撤单接口
    - other 其他接口
        - AutoInvest 定投接口
        - BsWap 币安挖矿接口
        - C2C C2C 接口
        - Convert 闪兑接口
        - CryptoLoans 质押借币接口
        - Fiat 法币接口
        - Futures 合约接口
        - FuturesAlgo 合约策略交易接口
        - GiftCard 币安礼品卡接口
        - Mining 矿池接口
        - Nft NFT 接口
        - Pay Pay 接口
        - Portfolio 经典统一账户接口
        - Rebate 返佣接口
        - SimpleEarn 赚币接口
        - SubAccount 子母账户接口
        - VipLoans VIP借币接口
        - Wallet 钱包接口
- app 常用接口与便捷功能封装
    - account（账户APP）
        - AccountSPOT
        - AccountUM
        - AccountCM
    - market（行情APP）
        - MarketSPOT
        - MarketUM
        - MarketCM
    - trade（便捷交易APP，支持异步下单，挂单超时取消等）
        - TradeSPOT
        - TradeUM
        - TradeCM
    - candle_server（K线服务，支持实时多产品K线异步维护）
        - CandleServer
        - CandleRule
