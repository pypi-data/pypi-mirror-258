```python
import ogr_rust as ogr
```

```python
rulers = ogr.enumerate_rulers(15)
```

```python
len(rulers)
```

```python
rulers = ogr.enumerate_rulers(20)
```

```python
print(len(rulers))
2 ** 20
```

```python
golomb = [r for r in rulers if r.is_golomb_ruler()]
```

```python
print(len(golomb))
golomb_ids = [r.to_id() for r in golomb]
```

```python
print(golomb_ids)
```

```python
let_orders = [r.order() for r in golomb]
```

```python
import polars as pl
df_orders = pl.DataFrame(dict(id=golomb_ids, order=let_orders))
df_orders.write_csv("first_2k.csv")
```

```python
import altair as alt
import polars as pl
import pandas as pd

# Show the histogram of numbers
data = dict(
    g_id=golomb_ids
)

df = pd.DataFrame(data)
```

```python
print(df)
alt.Chart(df).mark_rule().encode(
    x="g_id"
)
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()

x = df["g_id"]

# plt.plot(x, 10 * np.ones(np.shape(x)), '.')

plt.hist(x)

df.to_csv("first_2k.csv", index=False)
```
